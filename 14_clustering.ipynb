K-means clustering

```python
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)
centers = kmeans.cluster_centers_
print(centers)
plt.scatter(x, y, c=kmeans.labels_)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
X = np.random.normal(size =(1000 , 2)) # 1000 pts in 2 dims
print(np.shape(X))

clf = KMeans(n_clusters =2)
clf.fit(X)
centers = clf.cluster_centers_ # locations of the clusters
labels = clf.predict(X) # labels for each of the points
print(centers)
print(labels)
```

Mean shift is implemented in Scikit-learn

```python
import numpy as np
from sklearn.cluster import MeanShift
X = np.random.normal(size =(1000 , 2)) # 1000 pts in 2 dims
# if no bandwidth is specified , it will be learned from data
ms = MeanShift(bandwidth =1.0)
ms.fit(X) # fit the data
centers = ms.cluster_centers_ # centers of clusters
labels = ms.labels_ # labels of each point X

print(centers)
print(labels)
```

**Hierarchical Clustering**


The result is that T is a 10001000 sparse matrix, with T[i, j] = 0 for nonconnected points, and T[i, j] is the distance between points i and j for connected points. This algorithm will be efficient for small n_neighbors. If we set n_neighbors=1000 in this case, then the approximation to the Euclidean minimum spanning tree will be exact. For well-behaved data sets, this approximation is often exact even for k << N.

```python
from scipy.sparse.csgraph import minimum_spanning_tree
from sklearn.neighbors import kneighbors_graph
X = np.random.random ((1000 , 2)) # 1000 pts in 2 dims
G = kneighbors_graph(X, n_neighbors =10, mode='distance')
T = minimum_spanning_tree(G)

#print(G)
print(T)
```
